# 8: Create a new column group

We would like to split the class into two group : A and B.
If the first letter of the first name of the person is between A-L then this person group is A otherwise it's B

I give you the function that match if a character is in the range a-l (case insensitive)

```python
import re
def match(c):
    group = re.compile("[a-lA-L]")
    if group.match(c):
        return True
    return False
```

```python
L= []
for x in df2.index : 
  L.append(match(x))

df2['group'] = L
group1 = df2[df2['group']==True]    # Group 1 with label true / we can drop the column group juste après
group2 = df2[df2['group']==False]

group1
```
