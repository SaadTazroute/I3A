## t-SNE and LLE



"Why doesnâ€™t t-SNE work as well as LLE on the Swiss roll data?" from t-SNE FAQ -> try swiss roll for case where t-SNE is worse. 
"While Isomap, LLE and variants are best suited to unfold a single continuous low dimensional manifold, t-SNE will focus on the local structure of the data and will tend to extract clustered local groups of samples as highlighted on the S-curve example. This ability to group samples based on the local structure might be beneficial to visually disentangle a dataset that comprises several manifolds at once as is the case in the digits dataset."

```python
import matplotlib.pyplot as plt



def my_plot_2d_scatter(Xdata, Ydata, title="", colors=[]):
    
    fig = plt.figure(figsize=(12, 8))
    plt.scatter(Xdata, Ydata, c=colors, cmap=plt.cm.Spectral)
    plt.title(title)
    plt.show()
```

```python
from sklearn import manifold, datasets
from mpl_toolkits.mplot3d import axes3d, Axes3D

n_points = 1000
# X, color = datasets.samples_generator.make_s_curve(n_points, random_state=0)
X, color = datasets.make_swiss_roll(n_samples=n_points, random_state=0)
n_neighbors = 10
n_components = 2

fig = plt.figure(figsize=(12, 8))
# ax = fig.add_subplot(111, projection='3d')
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
# ax.view_init(4, -72)
plt.show()
print(color.shape)
```

```python
mds = manifold.MDS(n_components, max_iter=100, n_init=1)
Xmds = mds.fit_transform(X)
my_plot_2d_scatter(Xmds[:,0], Xmds[:,1], title="MDS", colors=color)

isomap = manifold.Isomap(n_neighbors, n_components)
Xiso = isomap.fit_transform(X)
my_plot_2d_scatter(Xiso[:,0], Xiso[:,1], title="Isomap", colors=color)

lle = manifold.LocallyLinearEmbedding(n_neighbors, n_components)
Xlle = lle.fit_transform(X)
my_plot_2d_scatter(Xlle[:,0], Xlle[:,1], title="LLE", colors=color)

tsne = manifold.TSNE(n_components=n_components, init='pca', random_state=0)
Xtsne = tsne.fit_transform(X)
my_plot_2d_scatter(Xtsne[:,0], Xtsne[:,1], title="t-SNE", colors=color)
```

Parmi les mÃ©thodes utilisÃ©es pour rÃ©duire la dimension, la mÃ©thode de T-sne reste la plus performante en termes de sÃ©parabilitÃ© des points sur l'espace 2D pour le dataset Mnist, un simple classifier pourra prendre le relais pour faire la classification aisÃ©ment.

```python
n_neighbors = 10
n_components = 2
```

```python
from time import time


def complexityMDS(L,X,y): # function permet de tracer la courbe de temps de calcul pour les diffÃ©rents nb_samples prÃ©sents sur la liste L pour l'algo MDS

  Timing = []
  for i in range(len(L)):
    mds = manifold.MDS(2, max_iter=100, n_init=1)

    start = time()
    print('processing' , L[i] , 'samples')

    X_sample , y_sample = X[:L[i],:] , y[:L[i]]
    Xmds = mds.fit_transform(X_sample)
    delta  = time()- start # temps de fitting sur 
    Timing.append(delta)
  fig = plt.figure(figsize=(12, 8))
  plt.plot(L , Timing) # tracÃ© de temps de calucl en fct de taille d'Ã©chantillons pris
  plt.title("Temps de calcul en fonction de nombre d'Ã©chantillons -MDS- ")
  plt.show()

  return L , Timing
```

```python

def complexityLLE(L,X,y):   # function permet de tracer la courbe de temps de calcul pour les diffÃ©rents nb_samples prÃ©sents sur la liste L pour LLE

  Timing = []
  for i in range(len(L)):
    lle = manifold.LocallyLinearEmbedding(n_neighbors, n_components)

    start = time()
    print('processing' , L[i] , 'samples')

    X_sample , y_sample = X[:L[i],:] , y[:L[i]]
    Xlle = lle.fit_transform(X_sample)    
    delta  = time()- start # temps de fitting sur 
    Timing.append(delta)
  fig = plt.figure(figsize=(12, 8))
  plt.plot(L , Timing) # tracÃ© de temps de calucl en fct de taille d'Ã©chantillons pris
  plt.title("Temps de calcul en fonction de nombre d'Ã©chantillons -LLE- ")
  plt.show()

  return L , Timing
```

```python

def complexityTSNE(L,X,y):  # function permet de tracer la courbe de temps de calcul pour les diffÃ©rents nb_samples prÃ©sents sur la liste L pour TSNE

  Timing = []
  for i in range(len(L)):
    tsne = manifold.TSNE(n_components=n_components, init='pca', random_state=0)

    start = time()
    print('processing' , L[i] , 'samples')

    X_sample , y_sample = X[:L[i],:] , y[:L[i]]
    Xtsne = tsne.fit_transform(X_sample)    
    delta  = time()- start # temps de fitting sur 
    Timing.append(delta)
  fig = plt.figure(figsize=(12, 8))
  plt.plot(L , Timing) # tracÃ© de temps de calucl en fct de taille d'Ã©chantillons pris
  plt.title("Temps de calcul en fonction de nombre d'Ã©chantillons -TSNE- ")
  plt.show()

  return L , Timing
```

```python

from time import time
def complexityIsomap(L,X,y):

  Timing = []
  for i in range(len(L)):
    isomap = manifold.Isomap(n_neighbors, n_components)

    start = time()
    print('processing' , L[i] , 'samples')

    X_sample , y_sample = X[:L[i],:] , y[:L[i]]
    Xisomap = isomap.fit_transform(X_sample)    
    delta  = time()- start # temps de fitting sur 
    Timing.append(delta)
  fig = plt.figure(figsize=(12, 8))
  plt.plot(L , Timing) # tracÃ© de temps de calucl en fct de taille d'Ã©chantillons pris
  plt.title("Temps de calcul en fonction de nombre d'Ã©chantillons -Isomap- ")
  plt.show()

  return L , Timing
```

```python
nb_samples = list(np.arange(500, 5000, 500)) # j'ai pas pu tourner la fonction complexity avec le nb de samples proposÃ©s (50 000 ) alors j'ai fait de petite valeurs et bien rÃ©parties 
A,b = complexityMDS(nb_samples,X_pca,y)
A,b = complexityLLE(nb_samples,X_pca,y)
A,b = complexityTSNE(nb_samples,X_pca,y)
A,b = complexityIsomap(nb_samples,X_pca,y)
```
